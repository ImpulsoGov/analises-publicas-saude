```python
import pandas as pd
import requests
```

## Download de número de equipes via requests

### Por municípios

```python
# Fazendo o download do arquivo
url = 'http://tabnet.datasus.gov.br/csv/cnes_cnv_equipebr163447187_19_163_35.csv' # link disponível no "copia .csv" após consulta no tabnet
#NOTE: o link é específico dessa consulta, para novas consultas, mudar o link

response = requests.get(url)

# Decodificando o conteúdo para a codificação correta
content = response.content.decode('ISO-8859-1')  # Ajuste conforme necessário

# Salvando o arquivo corrigido
with open('data/raw/quantitativo_equipes_mun_2026.csv', 'w', encoding='utf-8') as f:
    f.write(content)
```

## Leitura dos dados

```python
# Fazendo o download do arquivo

equipes = pd.read_csv('data/raw/quantitativo_equipes_mun_2026.csv', sep=";")

equipes

```

```python
equipes[['codigo_ibge', 'municipio']] = equipes['Município'].str.split(' ', expand=True, n=1)

equipes
```

```python
colunas_numericas = equipes.columns.difference(['codigo_ibge', 'municipio'])

equipes[colunas_numericas] = (
    equipes[colunas_numericas]
    .replace("-", '0')
    # .apply(pd.to_numeric, errors='coerce')
    # .astype(int)  # inteiro com suporte a NaN
)

equipes
```
