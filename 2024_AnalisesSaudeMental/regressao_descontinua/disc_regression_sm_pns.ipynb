```python
import pandas as pd
```

```python
pns = pd.read_csv('data/pns2019.csv', usecols=['V0001','V00291','V0031', 'P05402', 'P03201', 'I00102',
                                               'N012','N016','N010','N011','N014',
                                               'N017','N013','N015','N018', 'C008', 'UPA_PNS']
                                               )
pns
```

```python
# Filtering people living in state capitals or metropolitan regions
pns = pns.query('V0031 == 1 | V0031==2')

#Older than 17 years
pns = pns.query('C008 >= 18')

# Do not have healthcare
pns = pns.query('I00102 != 1')

# Non-abusive consumption of cigars
pns = pns.query('P05402 < 20 | P05402.isnull()')

# Non-abuvise comsumption of alcohol
pns = pns.query('P03201 != 1')

pns
```

```python
# # of people in population of interest
treatable_population = sum(pns['V00291'].dropna())
f"# of people in population of interest: {round(treatable_population):,}"
```

```python
# Removing rows with no records related to mental health
phq_columns = ['N012','N016','N010','N011','N014','N017','N013','N015','N018']
pns = pns.dropna(subset=phq_columns, how = 'all')
pns
```

```python
# PHQ-9 score calculation and classification (PHQ >= 10)
# Transforming PNS mental health variables scale (1-4) to PHQ-9 scale (0-3)
pns.loc[:,'total_phq'] = (pns[phq_columns]-1).sum(axis=1)
# pns['target'] = (pns['total_phq'] >= 10).astype(int)

pns.loc[:,"intervention_target"] = pns.loc[:,"total_phq"].apply(lambda x: 1 if (x >= 5 and x < 20) else 0)

pns
```

```python
threshold_population_4 = sum(pns.query('total_phq == 4')['V00291'])
threshold_population_5 = sum(pns.query('total_phq == 5')['V00291'])


print(
f"""# of people in threshold group 1 (PHQ-9 score = 4): {round(threshold_population_4):,}
# of people in threshold group 2 (PHQ-9 score = 5): {round(threshold_population_5):,}"""
)
```

```python
target_population = sum(pns.query('intervention_target == 1')['V00291'])


f"# of people in target group (PHQ-9 score between 5 and 19): {round(target_population):,} ({round(target_population/treatable_population, 4):.2%} of elegible population)"
```

```python
print(
f"""% of people in threshold group 1 (PHQ-9 score = 4) given the treatable group: {round(threshold_population_4/treatable_population, 4):.2%}
% of people in threshold group 2 (PHQ-9 score = 5) given the treatable group: {round(threshold_population_5/ treatable_population, 4):.2%}
---------------
% of people in threshold group 1 (PHQ-9 score = 4) given the target group: {round(threshold_population_4/target_population, 4):.2%}
% of people in threshold group 2 (PHQ-9 score = 5) given the target group: {round(threshold_population_5/ target_population, 4):.2%}"""
)
```

```python
f"% of individuals with close to PHQ-9 score threshold given target group: {round((threshold_population_4+threshold_population_5)/target_population, 4):.2%}"
```
