```python
import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
```

```python
#NOTE: due to the PNS file size (~400mb)

# Read PNS data and select variables of interest
pns = pd.read_csv('data/pns2019.csv', usecols=['B001','V00291','Q092', 'Q09201', 'Q09202',
                                               'N012','N016','N010','N011','N014',
                                               'N017','N013','N015','N018', 'C008']
                                               )
pns
```

```python
# IBGE projection
total_population_2019 = 210147125
adult_population_2019 = 156387668
```

```python
# Selecting individuals of legal age who are registered with the APS (Primary Health Care) 
pns = pns.query('B001 == 1 & C008 >= 18')
pns
```

```python
# # of adults registered with APS
registered_adults = sum(pns['V00291'].dropna())
registered_adults
```

```python
# Filtering people who do not take medication for depression and do not have a medical prescription nor a diagnosis

pns = pns.query('Q092 == 2 & Q09202 != 1 & Q09201 != 1')
pns
```

```python
# Removing rows with no records related to mental health
phq_columns = ['N012','N016','N010','N011','N014','N017','N013','N015','N018']
pns = pns.dropna(subset=phq_columns, how = 'all')
pns
```

```python
# PHQ-9 score calculation and classification (PHQ >= 10)
# Transforming PNS mental health variables scale (1-4) to PHQ-9 scale (0-3)
pns['total_phq'] = (pns[phq_columns]-1).sum(axis=1)
pns['target'] = (pns['total_phq'] >= 10).astype(int)
pns
```

```python
# Number of people with depression symptoms
total_depressivos = sum(pns['V00291'] * pns['target'])

# Number of people in the potential audience (adults within the APS)
total_amostra = sum(pns['V00291'])
```

```python
# Percent of individuals in the audience with PHQ-9 higher than 9

print(f"""
Total population in 2020: {total_population_2019:,}
Adult population in 2020: {adult_population_2019:,} ({(adult_population_2019/total_population_2019):.2%} of total population)
Adults registered with APS: {round(registered_adults):,} ({(registered_adults/total_population_2019):.2%} of total population)
Number of adults in APS with depression symptoms, undiagnosed (PHQ >= 10): {round(total_depressivos):,} ({(total_depressivos/total_population_2019):.2%} of total population)
""")
```

```python
#Proportion of individuals in APS with registered phones
prop_cadastrados = 24/100

## Error rates
numero_invalido = 2.6/100
numero_trocado = 3.6/100
```

```python
# Estimates of adults with depressions symptoms that could be reached by the program, given registered phones rates and sending errors
potentially_reached_adults = total_depressivos * prop_cadastrados * (1-numero_invalido+numero_trocado)

f"Potentially reached adults: {round(potentially_reached_adults):,} ({(potentially_reached_adults/total_population_2019):.2%} of total population)"
```

```python
# Proportion of individuals who, upon receiving a message, change their behavior (possibly going to the health unit to get the medication)
# Estimates of other types of health conditions were used, as described in the report

baseline_citopatologico = (40/1514)
rr_citopatologico = (47/1514)/baseline_citopatologico

baseline_cronicas = (264/1274)
rr_cronicas = (267/1273)/baseline_cronicas

baseline_vacinacao = (161/1000)
rr_vacinacao = (208/1000)/baseline_vacinacao

# Geometric mean
rr_media_geom_intervencoes = (rr_citopatologico * rr_cronicas * rr_vacinacao)**(1/3)

rr_media_geom_intervencoes
```

```python
# Estimated treatment effect in the control group

baseline_depressao = 30.8/100

mudanca_tratamento = (rr_media_geom_intervencoes * baseline_depressao) - baseline_depressao
mudanca_tratamento
```

```python
# Estimated total number of people in the interest group who would seek the medicine upon receiving the message (rounded down)
publico_atingido = total_depressivos * prop_cadastrados * (1-numero_invalido-numero_trocado) * mudanca_tratamento

print(f"""Estimativa do público atingido: {round(publico_atingido):,}
Estimativa do percentual atingido dentre a população geral: {(publico_atingido/total_population_2019):.2%}
Estimativa do percentual atingido dentre a população adulta: {(publico_atingido/adult_population_2019):.2%}
""")
```
